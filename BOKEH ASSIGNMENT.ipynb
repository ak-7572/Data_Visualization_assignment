```python
# Q1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

x = [1,2,3,4,5]
y = np.sin(x)
p = figure(title = 'sine wave', x_axis_label = "X", y_axis_label = "sin(x)")
p.line(x,y)
show(p)
```

```python
# Q2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

np.random.seed(42)
n = 100
x = np.random.rand(n) * 100
y = np.random.rand(n) * 100

sizes = np.random.rand(n) * 20 + 5
colors = np.random.choice(['red', 'blue', 'green', 'orange', 'purple'], n)

data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

source = ColumnDataSource(data)

p = figure(title="Scatter Plot with Varying Sizes and Colors", x_axis_label='X', y_axis_label='Y', width=800, height=400)

p.scatter('x', 'y', source=source, size='sizes', color='colors', fill_alpha=0.6)

output_notebook()
show(p)
```

```python
# Q3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.


fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts', width=600, height=400)

p.vbar(x='fruits', top='counts', source=source, width=0.4, color='navy', alpha=0.7)

output_notebook()
show(p)
```

```python
# Q4. Create a Bokeh histogram to visualize the distribution of the given data.

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

x = edges[:-1]
y = hist

p = figure(title="Histogram of Random Data",
           x_axis_label='Value',
           y_axis_label='Count',
           width=800,
           height=400)

p.vbar(x=x, top=y, width=0.8, color='blue', alpha=0.7)

output_notebook()
show(p)
```

```python
# Q5. Create a Bokeh heatmap using the provided dataset.

import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar
from bokeh.layouts import column

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)


x_flat = xx.flatten()
y_flat = yy.flatten()
z_flat = data_heatmap.flatten()

mapper = linear_cmap(field_name='z', palette='Viridis256', low=min(z_flat), high=max(z_flat))

source = {
    'x': x_flat,
    'y': y_flat,
    'z': z_flat
}

p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           width=600, height=600, tools='hover', tooltips=[("Value", "@z")])

p.rect(x='x', y='y', width=0.1, height=0.1, source=source, line_color=None,
        fill_color=mapper)

color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0, 0))
p.add_layout(color_bar, 'right')

output_notebook()
show(p)
```
