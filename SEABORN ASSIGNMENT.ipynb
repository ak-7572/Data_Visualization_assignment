```python
# Q2. Generate a dataset of random numbers. Visualize the distribution of a numerical variable.

import seaborn as sns
import pandas as pd
import numpy as np
length = np.random.rand(5)
breath = np.random.rand(5)

df = pd.DataFrame({'length' : length, 'breath' : breath})
df
```

```python
sns.relplot(x = 'length', y = 'breath', data = df, kind = 'line')
```
